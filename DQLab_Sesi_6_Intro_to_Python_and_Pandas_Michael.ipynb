# List / Array

```python
#             0         1        2      3     4     5
#             -6        -5      -4      -3    -2    -1
veggies = ["Tomat", "Wortel", "Kubis", 123, 89.0] #Kol
print(veggies[0]) # Output: Tomat
veggies.append("Kol") # add new item
print(veggies)
print(veggies[3])
print(veggies[-1]) # Output: Kol
print(veggies[1:3]) # Output: ['Wortel', 'Kubis']
print(veggies[1:]) # Output: ['Wortel', 'Kubis', 123, 89.0, 'Kol']
print(veggies[:2]) # Output: ['Tomat', 'Wortel', 'Kubis']
print(veggies[:-3]) # Output: ['Tomat', 'Wortel', 'Kubis', 123, 89.0]
print(veggies[0:5:2]) # Output: ['Tomat', 'Kubis', 'Kol']
print(veggies[::-1]) # Output: ['Kol', 89.0, 123, 'Kubis', 'Wortel', 'Tomat']
print(veggies[2::-1]) # Output: ['Kubis', 'Wortel', 'Tomat']
print(veggies[2:0:-1]) # Output: ['Kubis', 'Wortel']
```
