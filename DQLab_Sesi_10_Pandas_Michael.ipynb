# Filtering (Conditional Statement)

## Classic Style

```python
df[df['sales'] > 1000].head()
```

## Query Style

```python
# Query
df.query('sales > 1000').head()
```

# Multiple Conditions

## Classic Style

```python
# menyaring dengan dua atau lebih kondisi
# Note: untuk menghindari penulisan penulisan ekspresi, gunakan tanda kurung

# Gunakan bitwise operation
# & untuk AND
# | untuk OR
# ~ untuk NOT

# memfilter transaksi dengan sales > 100 dari kategori Furniture

df[(df['sales'] > 1000) & (df['quantity'] > 5) & (df['category'] == 'Furniture')].head()
```

```python
df[(df['sales'] > 1000) | (df['quantity'] > 5) | (df['state'] == 'California')].head()
```

```python
df[~(df['category'] == 'Furniture')].head()
```

## Query Style

```python
df.query('sales > 1000 and quantity > 5').head()
```

```python
df.query('sales > 1000 or quantity > 5 or state == "California"').head()
```

```python
df.query('not category == "Furniture"').head()
```

```python
def penjumlahan(a,b):
  return a+b
```

```python
def pengurangan(a,b):
  return a-b
```

```python
# Positional Argument
penjumlahan(5,5)
```

```python
pengurangan(5,5)
```

```python
# Keyword Argument
penjumlahan(a=5,b=5)
```

```python
pengurangan(b=5,a=5)
```

```python
# notes: positional dulu, baru keyword
penjumlahan(5,b=5)
```

```python
# notes: positional dulu, baru keyword
penjumlahan(a=5,5)
```

# Summarizing Data

```python
df
```

```python
summarize_df = df.groupby(['order_date','region','category']).agg({'sales':'sum', 'quantity':'sum'})
summarize_df
```

```python
summarize_df.reset_index()
# Untuk menyamakan level dari sebuah kolom data, namun tidak mengubahbentuk data summarize_df awal
```

```python
summarize_df.reset_index(inplace=True)
#View Table tidak dilihatkan tanpa memanggilnya lagi, dan disave untuk format barunya
```

```python
summarize_df
```

*Kalau tak mau mengubah reset index dan di-inplace, ulangi dari step teratas dengan tabel yang belum berubah bentuk*
